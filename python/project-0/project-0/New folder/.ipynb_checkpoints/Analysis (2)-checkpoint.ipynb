# Imports

```python
%pip install matplotlib
%pip install pandas
%pip install nltk
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from collections import Counter
from sklearn.metrics import silhouette_score
import numpy as np

```

# Start
First we read the data from the csv file and show the first few rows to examine the data.

```python

# Load the CSV file
file_path = './all_sources_metadata_2020-03-13.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
data.head()
```


The DataFrame consists of columns like 'title', 'doi', 'abstract', 'publish_time', 'authors', and 'journal'. Some columns have missing values, such as 'abstract', 'pmcid' and 'pubmed_id'.

# Preprocessing
We'll clean the data by handling missing values, particularly in the 'title' and 'abstract' columns, which are crucial for our analysis.

```python
# Preprocessing
# Removing rows where either title or abstract is missing
data_cleaned = data.dropna(subset=['title', 'abstract']).copy()  # Added .copy() to explicitly create a copy of the DataFrame

# Converting to lowercase for uniformity
data_cleaned.loc[:, 'title'] = data_cleaned['title'].str.lower()
data_cleaned.loc[:, 'abstract'] = data_cleaned['abstract'].str.lower()

# Display the first few rows of the cleaned DataFrame
data_cleaned.head()
```

```python
# Removing Stop words

nltk.download('stopwords')
stop = stopwords.words('english')

data_cleaned['title'] = data_cleaned['title'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data_cleaned['abstract'] = data_cleaned['abstract'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

We removed the stop words from the title and abstract columns. Let's see how the first few rows look like now.

```python
data_cleaned.head()
```

# Tokenization
We then split the text into individual words or tokens


```python
nltk.download('punkt')

data_cleaned['title'] = data_cleaned['title'].apply(word_tokenize)
data_cleaned['abstract'] = data_cleaned['abstract'].apply(word_tokenize)
```

# Stemming 
we reduce the inflected words to their word stem or root form!

```python
# Stemming
porter = PorterStemmer()

data_cleaned['title'] = data_cleaned['title'].apply(lambda x: [porter.stem(word) for word in x])
data_cleaned['abstract'] = data_cleaned['abstract'].apply(lambda x: [porter.stem(word) for word in x])

data_cleaned.head()
```

# Feature Extraction Using TF-IDF
Now that the data is cleaned, we will extract features from the 'abstract' column using the TF-IDF (Term Frequency-Inverse Document Frequency) method. This will transform the text into a numerical format suitable for topic modeling and clustering.

```python

# Convert the elements of the 'abstract' column back to strings
data_cleaned['abstract'] = data_cleaned['abstract'].apply(lambda x: ' '.join(x))

# Initializing the TF-IDF vectorizer with a maximum of 1000 features
tfidf_vectorizer = TfidfVectorizer(max_features=1000)

# Transforming the abstracts
tfidf_matrix = tfidf_vectorizer.fit_transform(data_cleaned['abstract'])

data_cleaned['abstract_tdidf'] = list(tfidf_matrix.toarray())

data_cleaned.head()
```

# Feature extraction using Bag of Words

```python

# Create a CountVectorizer object
count_vectorizer = CountVectorizer()

# Fit and transform the abstracts using the CountVectorizer
abstract_bow = count_vectorizer.fit_transform(data_cleaned['abstract'])

# Convert the bag of words representation into a matrix
bow_matrix = abstract_bow.toarray()

# Print the shape of the matrix to check its dimensions
print(bow_matrix.shape)

# Display the first few rows of the DataFrame
data_cleaned.head()
```

# Topic Modeling - LDA
here I chose 10 topics to extract from the data and printed them.
I trained the model 2 times for each of the feature extraction models I used earlier.

### tfidf

```python

# Now you can get the feature names
feature_names = tfidf_vectorizer.get_feature_names_out()
# Number of topics to identify
n_topics = 10

# Initializing LDA model
lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=0)

# Fitting the model to the TF-IDF matrix
lda_model.fit(tfidf_matrix)

# Create a function to display the most important words for each topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Display the most important words for each topic
no_top_words = 10
display_topics(lda_model, tfidf_vectorizer.get_feature_names_out(), no_top_words)
```

### BoW

```python
# Create and fit the LDA model on the bow_matrix
lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=0)
lda_model.fit(bow_matrix)

# Get the topics
topics = lda_model.transform(bow_matrix)

# Print the shape of the topics to check its dimensions
print(topics.shape)
```

```python
# Number of top words to display for each topic
no_top_words = 10

# Getting feature names
feature_names = count_vectorizer.get_feature_names_out()

# Display the topics
display_topics(lda_model, feature_names, no_top_words)
```

# Dimensionality Reduction

I used TruncatedSVD instead of the mentioned methods in the Project Document because some of them weren't for sparse matrices.

```python
# Number of components for SVD
n_components = 2

# Apply TruncatedSVD for dimensionality reduction on TF-IDF features
svd = TruncatedSVD(n_components=n_components)
tfidf_svd = svd.fit_transform(tfidf_matrix)

# And then apply TruncatedSVD for dimensionality reduction on bag of words features
svd = TruncatedSVD(n_components=n_components)
bow_svd = svd.fit_transform(bow_matrix)
```

# Elbow method

Using elbow method to find the best number of clusters.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming tfidf_svd and bow_svd are your dimensionality-reduced feature sets
feature_sets = {'TF-IDF': tfidf_svd, 'BoW': bow_svd}
elbow_data = {}

for name, features in feature_sets.items():
    wcss = []
    for i in range(1, 11):  # Let's test from 1 to 10 clusters
        kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
        kmeans.fit(features)
        wcss.append(kmeans.inertia_)
    elbow_data[name] = wcss

    # Plot the WCSS to see the 'elbow'
    plt.plot(range(1, 11), wcss, marker='o')
    plt.title(f'Elbow Method for {name} Features')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()
```

It seems like the 3 would be a good choice for both of them

```python
n_clusters = 3
```

# Clustering

### K-means Tfidf

```python
# Initialize K-means with n_init=10
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init=10)

# Fit the model to the TF-IDF matrix
kmeans.fit(tfidf_svd)

clusters_tfidf = kmeans.predict(tfidf_svd)

# Print the cluster labels
print(clusters_tfidf)
```

Let's see how many entries for each cluster in tfidf

```python
# Count the occurrences of each unique cluster
cluster_counts = Counter(clusters_tfidf)

# Print the cluster counts
print(cluster_counts)
```

```python
# Add the cluster labels to the DataFrame
data_cleaned['cluster_tfidf'] = clusters_tfidf

# Display the first few rows with cluster labels
data_cleaned.head()
```

### BoW

here we fit kmeans to bow, but before that we remove outliers (there were some weird outliers that would affect the final results)

```python
# Calculate the interquartile range for feature 1 and feature 2
Q1 = np.percentile(bow_svd, 25, axis=0)
Q3 = np.percentile(bow_svd, 75, axis=0)
IQR = Q3 - Q1

# Define outlier thresholds (1.5 times IQR away from Q1 and Q3)
feature1_min = Q1[0] - 1.5 * IQR[0]
feature1_max = Q3[0] + 1.5 * IQR[0]
feature2_min = Q1[1] - 1.5 * IQR[1]
feature2_max = Q3[1] + 1.5 * IQR[1]

# Filter out outliers
mask = np.all(np.logical_and(bow_svd >= [feature1_min, feature2_min], 
                             bow_svd <= [feature1_max, feature2_max]), axis=1)
bow_svd = bow_svd[mask]

# Fit the model to the BOW matrix
kmeans.fit(bow_svd)
clusters_bow = kmeans.predict(bow_svd)

# Print the cluster labels
print(clusters_bow)
```

```python
# Count the occurrences of each unique cluster
cluster_counts = Counter(clusters_bow)

# Print the cluster counts
print(cluster_counts)
```

It's less normally distributed compared to tfidf

## Silhouette score for K-means

```python
# Evaluate clusters with silhouette score
silhouette_avg_tfidf = silhouette_score(tfidf_svd, clusters_tfidf)
print("Silhouette Score(tfidf): ", silhouette_avg_tfidf)

silhouette_avg_bow = silhouette_score(bow_svd, clusters_bow)
print("Silhouette Score(bow): ", silhouette_avg_bow)
```

BoW have more distinct clusters than tfidf

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

feature_sets = {'TF-IDF': tfidf_svd, 'BoW': bow_svd}

for name, features in feature_sets.items():
    n_clusters = 3  

    # Apply Agglomerative Clustering
    hc = AgglomerativeClustering(n_clusters=n_clusters, metric='euclidean', linkage='ward')
    hc_clusters = hc.fit_predict(features)

    # Plot the clusters
    plt.scatter(features[:, 0], features[:, 1], c=hc_clusters, marker='o')
    plt.title(f'Clusters from Hierarchical Clustering on {name}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Remove outliers from the BoW matrix
bow_svd = bow_svd[np.all(bow_svd < [100, 100], axis=1)]

# Plot TF-IDF
axs[0].scatter(tfidf_svd[:, 0], tfidf_svd[:, 1], c=clusters_tfidf, s=50, cmap='viridis')
axs[0].set_xlabel('SVD Feature 1')
axs[0].set_ylabel('SVD Feature 2')
axs[0].set_title('2D Visualization of Clusters with TF-IDF')

# Plot BoW
axs[1].scatter(bow_svd[:, 0], bow_svd[:, 1], c=clusters_bow, s=50, cmap='viridis')
axs[1].set_xlabel('SVD Feature 1')
axs[1].set_ylabel('SVD Feature 2')
axs[1].set_title('2D Visualization of Clusters with BoW')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

# Plotting Altogether

```python
from sklearn.metrics import silhouette_score
# Create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Remove outliers from the BoW matrix
bow_svd = bow_svd[np.all(bow_svd < [100, 100], axis=1)]

# Plot TF-IDF
axs[0, 0].scatter(tfidf_svd[:, 0], tfidf_svd[:, 1], c=clusters_tfidf, s=50, cmap='viridis')
axs[0, 0].set_xlabel('Feature 1')
axs[0, 0].set_ylabel('Feature 2')
axs[0, 0].set_title('KMeans TF-IDF')

# Calculate and display the silhouette score for KMeans TF-IDF
silhouette_avg_tfidf = silhouette_score(tfidf_svd, clusters_tfidf)
axs[0, 0].text(0.05, 0.95, f'Silhouette Score: {silhouette_avg_tfidf:.2f}', transform=axs[0, 0].transAxes)

# Plot BoW
axs[0, 1].scatter(bow_svd[:, 0], bow_svd[:, 1], c=clusters_bow, s=50, cmap='viridis')
axs[0, 1].set_xlabel('Feature 1')
axs[0, 1].set_ylabel('Feature 2')
axs[0, 1].set_title('KMeans BoW')

# Calculate and display the silhouette score for KMeans BoW
silhouette_avg_bow = silhouette_score(bow_svd, clusters_bow)
axs[0, 1].text(0.05, 0.95, f'Silhouette Score: {silhouette_avg_bow:.2f}', transform=axs[0, 1].transAxes)

feature_sets = {'TF-IDF': tfidf_svd, 'BoW': bow_svd}

for i, (name, features) in enumerate(feature_sets.items()):
    n_clusters = 3  

    # Apply Agglomerative Clustering
    hc = AgglomerativeClustering(n_clusters=n_clusters, metric='euclidean', linkage='ward')
    hc_clusters = hc.fit_predict(features)

    # Plot the clusters
    axs[1, i].scatter(features[:, 0], features[:, 1], c=hc_clusters, marker='o')
    axs[1, i].set_title(f'Hierarchical Clustering {name}')
    axs[1, i].set_xlabel('Feature 1')
    axs[1, i].set_ylabel('Feature 2')

    # Calculate and display the silhouette score for Hierarchical Clustering
    silhouette_avg_hc = silhouette_score(features, hc_clusters)
    axs[1, i].text(0.05, 0.95, f'Silhouette Score: {silhouette_avg_hc:.2f}', transform=axs[1, i].transAxes)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Looking at the provided plots and silhouette scores:

- BoW with KMeans has the best score (0.47), meaning it grouped the data most clearly.
- TF-IDF with KMeans scored lower (0.40), which means its groups were less distinct.
- BoW also did slightly better with Hierarchical Clustering than TF-IDF.

BoW and KMeans together did the best job of creating clear groups from the data.
